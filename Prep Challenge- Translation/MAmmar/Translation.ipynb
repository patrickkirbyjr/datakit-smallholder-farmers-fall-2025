# Deep Learning Model Instantiated

```python
from transformers import pipeline

# Load Meta's NLLB model for translation
translator = pipeline(
    "translation",
    model="facebook/nllb-200-distilled-600M",
    tokenizer="facebook/nllb-200-distilled-600M"
)
```

# Translating first 100 rows after cleaning

```python
import re


# -------------------------------
# 2️⃣ Map short language codes to NLLB format
# NLLB-200 uses codes like <language_code>_<script>
# where 'Latn' indicates Latin alphabet (e.g. 'swh_Latn' = Swahili in Latin script)
# -------------------------------
lang_map = {
    "nyn": "nyn_Latn",   # Nyankole
    "swa": "swh_Latn",   # Swahili
    "lug": "lug_Latn",   # Luganda
    "eng": "eng_Latn"    # English (no translation)
}

# -------------------------------
# 3️⃣ Cleaning function to remove extra characters
# Removes question numbers, strange symbols, repeated letters etc.
# -------------------------------
def clean_text(text):
    if pd.isna(text):
        return text
    text = str(text)
    text = re.sub(r'Q\d+\s*', '', text)          # remove question numbers like Q1, Q23
    text = re.sub(r'[^\w\s?.!,-]', ' ', text)    # remove odd characters/symbols
    text = re.sub(r'\bS+\b', '', text)           # remove single or repeated 'S' noise
    text = re.sub(r'\s+', ' ', text).strip()     # clean extra spaces
    return text

# -------------------------------
# 4️⃣ Safe translation helper function
# Handles missing data and translation errors
# -------------------------------
def translate_text(text, src):
    try:
        if src not in lang_map:     # unknown language code
            return f"[error: unknown lang {src}]"
        if src == "eng" or pd.isna(text):
            return text  # skip translation for English or empty cells
        result = translator(text, src_lang=lang_map[src], tgt_lang="eng_Latn")
        return result[0]['translation_text']
    except Exception as e:
        return f"[error: {e}]"  # Handles errors safely during translation so the code keeps running even if one row fails

# -------------------------------
# 5️⃣ Take only the first 100 rows
# -------------------------------
subset = chunk.head(100).copy()

# Clean the text before translating
subset['question_content'] = subset['question_content'].apply(clean_text)
subset['response_content'] = subset['response_content'].apply(clean_text)

# -------------------------------
# 6️⃣ Apply translation
# -------------------------------
subset['question_translated'] = subset.apply(
    lambda row: translate_text(row['question_content'], row['question_language']), axis=1
)

subset['response_translated'] = subset.apply(
    lambda row: translate_text(row['response_content'], row['question_language']), axis=1
)

# -------------------------------
# 7️⃣ Display results
# -------------------------------
pd.set_option('display.max_rows', 100)
pd.set_option('display.max_colwidth', None)
subset[['question_translated', 'response_translated']]
```
